## Train/Validation Split

```python
import os
from tqdm import tqdm

# Set Inputs Here
voxelnet_folder = '../src/Voxelnet/'
kitti_folder = '../data/stage3/kitti/object/'


# Move files from training to validation
if not os.path.exists(kitti_folder):
    raise FileNotFoundError('Error: kitti_folder does not exist.')
if not os.path.exists(voxelnet_folder):
    raise FileNotFoundError('Error: voxelnet_folder does not exist.')

# os.mkdir, ignore if already exists
os.makedirs(kitti_folder + '/validation/image_2/', exist_ok=True)
os.makedirs(kitti_folder + '/validation/label_2/', exist_ok=True)
os.makedirs(kitti_folder + '/validation/velodyne_crop/', exist_ok=True)
os.makedirs(kitti_folder + '/validation/calib/', exist_ok=True)

lines_val = [line.rstrip('\n') for line in open(voxelnet_folder + '/crop_data/val.txt')]

# Move files
for frame_number in tqdm(lines_val):
    os.rename(f'{kitti_folder}/training/image_2/{frame_number}.png', f'{kitti_folder}/validation/image_2/{frame_number}.png')
    os.rename(f'{kitti_folder}/training/label_2/{frame_number}.txt', f'{kitti_folder}/validation/label_2/{frame_number}.txt')
    os.rename(f'{kitti_folder}/training/velodyne_crop/{frame_number}.bin', f'{kitti_folder}/validation/velodyne_crop/{frame_number}.bin')
    os.rename(f'{kitti_folder}/training/calib/{frame_number}.txt', f'{kitti_folder}/validation/calib/{frame_number}.txt')
```
